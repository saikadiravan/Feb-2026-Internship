Problem 1: Mobile Recharge Validation System

```python
valid_plans = [199, 299, 399, 599]

#Function to validate recharge
def validate_recharge(amount):
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False

    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans:", valid_plans)
        return False

    return True


#Retry logic using while loop
while True:
    amount = int(input("Enter recharge amount: "))

    if validate_recharge(amount):
        print("Recharge Successful!")
        break   # Exit loop if valid
    else:
        print("Please try again.\n")
```

Problem 2: Inventory Reorder Alert System

```python
products = {
    "Laptop": 25,
    "Mouse": 10,
    "Keyboard": 18,
    "Monitor": 11
}

#Function to check stock
def check_stock(product_dict):

    for product in product_dict:

        quantity = product_dict[product]

        if quantity < 15:
            print(product + ": Reorder Alert")
        else:
            print(product + ": Stock OK")


#Call the function
check_stock(products)
```

 Problem 3: Student Result Processing System

```python
def evaluate_student(marks):

    total = 0

    #Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

student_marks = [60, 55, 70, 45]

#Call the function
result = evaluate_student(student_marks)

print("Average:", sum(student_marks) / len(student_marks))
print("Result:", result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak_hour):

    base_fare = 50
    per_km_rate = 12

    total_fare = base_fare + (distance * per_km_rate)

    #Add 25% extra during peak hour
    if peak_hour == "yes":
        total_fare = total_fare + (total_fare * 0.25)

    return total_fare


#Retry using while loop
while True:

    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour(yes/no)?: ").lower()

    fare = calculate_fare(distance, peak)

    print("Total Fare: ₹", fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()

    if retry != "yes":
        print("Thank you for using cab fare calculator.")
        break
```

Problem 5: Employee Attendance Eligibility

```python
def check_attendance(attendance_list):

    present_count = 0

    #Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_count += 1

    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P","A","P","P","A"]

result = check_attendance(attendance)

print("Attendance Percentage:", (attendance.count("P") / len(attendance)) * 100)
print("Status:", result)
```

Problem 6: Password Strength Checker

```python
def check_password(password):

    #Rule:1- Minimum length
    if len(password) < 8:
        return "Weak Password (Minimum length is 8)"

    has_digit = 0
    has_special = 0

    #Loop through each character
    for char in password:

        if char.isdigit():
            has_digit = 1

        if char in "@#$":
            has_special = 1

    #Rule:2 Digit must be there
    if has_digit == 0:
        return "Weak Password (Must contain at least one digit)"
    #Rule:3 Special character must be there

    if has_special == 0:
        return "Weak Password (Must contain at least one special character @#$)"

    return "Strong Password"

password = input("Enter password: ")
result = check_password(password)

print(result)
```
