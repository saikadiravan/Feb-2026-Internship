```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
```

```python
model_name = "google/flan-t5-large"

tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

print("Model loaded successfully!")
```

```python
print("Chatbot: Hello! I am your AI assistant. Type 'exit' to quit.\n")

# Store last question for context
last_question = ""

while True:
    user_input = input("User: ")

    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye! 👋")
        break

    # Handle greetings
    if user_input.lower() in ["hi", "hello", "hey"]:
        print("Chatbot: Hello! How can I assist you today?")
        continue

    # Handle follow-up like "explain more"
    if user_input.lower() in ["explain more", "tell me more", "elaborate"]:
        if last_question != "":
            user_input = last_question
        else:
            print("Chatbot: Please ask something first.")
            continue

    # Save last meaningful question
    last_question = user_input

    # Strong instruction prompt
    prompt = f"""
Explain the following topic clearly in 3-4 sentences with an example:

{user_input}
"""
    # Tokenize
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)

    # Generate response (stable)
    outputs = model.generate(
        inputs["input_ids"],
        max_new_tokens=100,
        repetition_penalty=1.5,
        do_sample=False
    )

    # Decode response
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Safety fallback
    if response.strip() == "" or len(response) < 5:
        response = "I'm sorry, I couldn't understand that properly. Please try again."

    print(f"Chatbot: {response}")
```
