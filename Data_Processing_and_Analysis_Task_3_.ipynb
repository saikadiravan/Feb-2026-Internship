Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

#Find highest score
highest_score = max(employees.values())

#Find all employees with highest score, so use for loop
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print("Top performers eligible for bonus:",
      ", ".join(top_performers),
      "(score:", highest_score, ")")
```

Problem Statement 2: Search Query Keyword Analysis

```python
import string
query = "Buy mobile phone buy phone online"

#Convert to lowercase
query = query.lower()

#Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

#Split into words
words = query.split()

#Count frequency using dictionary
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

repeated_keywords = {}

for word, count in word_count.items():
    if count > 1:
        repeated_keywords[word] = count

print(repeated_keywords)
```

Problem Statement 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, reading in enumerate(sensor_readings):
  #Check if reading is even or not
    if reading % 2 == 0:
        valid_readings.append((index, reading))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

#Dictionary to store domain counts
domain_count = {}

#Count each domain
for email in emails:
    parts = email.split("@")   # Split using @
    domain = parts[1]          # Take domain part

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

#Calculate percentage
total = len(emails)

print("Email Domain Usage:")

for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain + ":", int(percentage), "%")
```

Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2300, 1400, 3000]

#Calculate average sales
total = 0
for value in sales:
    total += value

average = total / len(sales)

#Calculate 30% above average threshold
threshold = average * 1.3

#Detect spike days by for loop
for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day", i + 1, ":", sales[i])
```

Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

#Dictionary to count occurrences
count = {}

#Count each user ID through for loop
for user in user_ids:
    if user in count:
        count[user] += 1
    else:
        count[user] = 1

#Display duplicates only
for user in count:
    if count[user] > 1:
        print(user, "→", count[user], "times")
```
