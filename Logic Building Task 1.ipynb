1. User Login Check

```python
correct_username = "admin"
correct_password = "1234"

username = "admin"
password = "1234"

if username==correct_username and password==correct_password:
    print("Login successful")
else:
    print("Invalid credentials")
```

2. Pass / Fail Analyzer


```python
marks = [45, 78, 90, 33, 60]

pass_count=0
fail_count=0

for i in marks:
  if(i>=50):
    pass_count+=1
  else:
    fail_count+=1
print("Pass count:",pass_count)
print("Fail count:",fail_count)
```

3. Simple Data Cleaner


```python
names = [" Alice ", "bob", " CHARLIE "]

preprocessed_names = []

for i in names:
  preprocessed_name=i.strip()
  preprocessed_name=preprocessed_name.lower()
  preprocessed_names.append(preprocessed_name)


print(preprocessed_names)
```

4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for i in messages:
  length=len(i)
  print("message:",i)
  print("length:",length)
  if length>10:
    print("This message is longer than 10 characters")
```

5. Error Message Detector


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0
for i in logs:
    if i== "ERROR":
        error_count += 1

print(error_count)
```
