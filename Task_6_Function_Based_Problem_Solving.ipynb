Problem Statement 1:Smart Parking Lot Management System

```python
def manage_parking(capacity, logs):

    current_vehicles = 0
    peak_usage = 0

    for log in logs:

        if log == "IN":
            current_vehicles += 1

        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1

        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)

    #Check parking status
    if current_vehicles > capacity:
        print("Parking Status: Over Capacity Alert")
    else:
        print("Parking Status: Available")
    #This calculates the maximum number of vehicles at any time
    print("Peak Parking Usage:", peak_usage)


#Sample input
parking_capacity = 20
vehicle_logs = ["IN", "IN", "OUT", "OUT", "IN", "IN", "OUT","IN","IN","IN","IN","OUT"]

#Call the function
manage_parking(parking_capacity, vehicle_logs)
```

Problem Statement 2:Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):

    #Assume base time is 10 mins
    base_time = 15

    #Adjust time based on traffic
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 8

    #Adjust time based on weather
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20

    return round(base_time)

distance = float(input("Enter Distance (km): "))
traffic = input("Enter Traffic Level (Low/Medium/High): ")
weather = input("Enter Weather (Clear/Rainy/Stormy): ")

eta = estimate_delivery_time(distance, traffic, weather)

print("Estimated Delivery Time:", eta, "minutes")
```

Problem Statement 3:Movie Theatre Seat Occupancy

```python
def analyze_show(total_seats, booked_seats):

    booked_count = 0

    #Count booked seats using loop
    for seat in booked_seats:
        booked_count += 1

    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    #Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open additional show")

    elif occupancy >= 75:
        print("Show Status: Almost Full")

    else:
        print("Show Status: Seats Available")

#Sample Data
total_seats = 250
booked_seats = [1] * 229

analyze_show(total_seats, booked_seats)
```

 Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):

    total = 0

    #Calculate total CPU usage using for loop
    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    #Classify server status based on conditions
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")

#Sample Data
cpu_data = [40, 55, 60, 77, 80, 90]

classify_server_load(cpu_data)
```

 Problem Statement 5:Smart Classroom Resource Usage Monitor

```python
def analyze_resource_usage(resources):

    overused = []

    #Check each resource using for loop
    for resource in resources:
        hours = resources[resource]

        if hours > 8:   #Threshold for overuse
            overused.append(resource)

    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


#Sample Data
usage_data = {
    "Projector": 5,
    "AC": 10,
    "Lights": 20
}

analyze_resource_usage(usage_data)
```

 Problem Statement 6: Online Event Registration Capacity Controller

```python
def manage_registrations(capacity, total_registrations):

    confirmed = 0
    waitlisted = 0

    for i in range(total_registrations):

        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if waitlisted > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

#Sample Input
event_capacity = 500
registrations = 501

manage_registrations(event_capacity, registrations)
```
