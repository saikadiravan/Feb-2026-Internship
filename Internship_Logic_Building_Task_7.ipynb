Problem 1: Smart Light Controller

```python
class Smartlight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(self.name + " is " + self.status)

#Take input from user
light_name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ").upper()

#Create an object
light = Smartlight(light_name)

#Perform action
if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()

light.display_status()
```

Problem 2: Employee ID Card System

```python
class Employee:

    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\nEmployee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

#Take input from user
name = input("Enter Employee Name: ")
emp_id = input("Enter Employee ID: ")
department = input("Enter Department: ")

#Create an object
employee = Employee(name, emp_id, department)

employee.display_id_card()
```

Problem 3: Mobile Contact Record

```python
class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("\nContact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)

#Take input from user
name = input("Enter Contact Name: ")
phone = input("Enter Phone Number: ")

#Create an object
contact = Contact(name, phone)

contact.display_contact()
```

Problem 4: Product Price Tag Generator

```python
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print("\nProduct:", self.name)
        print("Price: ₹" + str(self.price))

#Sample input
product_name = "Sofa"
price = 5000

#Create an object
item = Product(product_name, price)
item.print_price_tag()
```

Problem 5: Movie Rating Display System

```python
class Movie:

    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print("\nMovie:", self.name)
        print("Rating:", self.rating, "/ 5")

#Take input from user
movie_name = input("Enter Movie Name: ")
rating = float(input("Enter Rating: "))

#Create an object
film = Movie(movie_name, rating)

film.display_details()
```

Problem 6: Delivery Address Manager

```python
class Delivery:

    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_details(self):
        print("\nDelivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)

#Take input from user
name = "Sai Kadiravan S"
address = "No. 45, 4th cross, 5th main road, JP Nagar"

#Create an object
order = Delivery(name, address)

order.print_details()
```
