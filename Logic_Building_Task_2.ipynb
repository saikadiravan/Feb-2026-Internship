 Problem 1: Unique Words in a Sentence

```python
sentence = input("Enter sentence:")
sentence = sentence.lower()

#Split sentence into words
words = sentence.split()

#Convert list to set to remove duplicates
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_salary = 0
highest_employee = ""

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)
```

Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

#Initialize max and min with first element
max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
#To store count of present days
present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

#Total number of days
total_days = len(attendance)

#Calculate percentage
attendance_percentage = (present_count / total_days) * 100

print("Attendance Percentage:", attendance_percentage)
```

Problem 7: Count Character Frequency

```python
text = "python"

# Empty dictionary to store counts
char_count = {}
#use for loop for looping through each character
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```

Problem 10: Calculate Average Marks

```python
marks = [65,74,83,92,40]
total = 0

for mark in marks:
    total += mark

# Calculate average
average = total / len(marks)

print("Average Marks:", average)
```
